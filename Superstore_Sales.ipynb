```python
import pandas as pd
url="https://raw.githubusercontent.com/anamikasharma20/Superstore-Sales-Dataset-Analysis/refs/heads/main/Superstore%20Sales%20Dataset.csv"
df=pd.read_csv(url)
df.head()
```

```python
import sqlite3

conn = sqlite3.connect("sales.db")
```

```python
df.to_sql("sales", conn, if_exists="replace", index=False)
```

```python
# SQL Query: Sales by Category

query = """
Select Sum(Sales) as Total_Sales from sales ;
"""

category_sales = pd.read_sql(query, conn)
category_sales
```

```python
# SQL Query: Sales by Region
query ="""
SELECT Region, SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Region
ORDER BY Total_Sales DESC;"""

region_sales = pd.read_sql(query, conn)
region_sales
```

```python
# SQL Query: Sales by Segment
query ="""SELECT Segment, SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Segment
ORDER BY Total_Sales DESC;"""
Segment_sales =pd.read_sql(query, conn)
Segment_sales
```

```python
# SQL Query: Top 10 Products by Sales

query = """
SELECT "Product Name", SUM(Sales) AS Total_Sales
FROM sales
GROUP BY "Product Name"
ORDER BY Total_Sales DESC
LIMIT 10;
"""

top_products = pd.read_sql(query, conn)
top_products
```

```python
# SQL Query: Orders Count by Region

query = """
SELECT Region, COUNT("Order ID") AS Total_Orders
FROM sales
GROUP BY Region;
"""

region_orders = pd.read_sql(query, conn)
region_orders.head()
```
